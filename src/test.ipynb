```python
import shutil
from pathlib import Path
from collections import Counter
```

```python
DIR_PATH = Path("/mnt/c/Users/ireal/Downloads")
```

```python
file_type_dest = {
    '.png':'image',
    '.jpg':'image',
    '.csv':'data',
    '.mp4':'video',
    '.zip':'compressed',
    '.ipynb':'notebook',
    '.json' : 'data' ,
    '.jpg' : 'image',
}
```

```python
file_extensions = []
for file_path in DIR_PATH.iterdir():
    #ignor directories
    if file_path.is_dir():
        continue
    
    # ignor hidden files
    if file_path.name.startswith('.'):
        continue
        
    # move file
    file_extensions.append(file_path.suffix)
    if file_path.suffix not in file_type_dest:
        DEST_DIR = DIR_PATH / 'other'
    else :
        DEST_DIR = DIR_PATH / file_type_dest [file_path.suffix]
        
    DEST_DIR.mkdir(exist_ok = True)
    print(f"{file_path.suffix:10} {DEST_DIR}")
    shutil.move(str(file_path) , str(DEST_DIR))
```
